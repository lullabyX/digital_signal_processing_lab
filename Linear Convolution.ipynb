# 02 Compute linear convolution of two signals. Compare the result with the output of the Matlab/Python library function.

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

## Signal Generation

```python
# impulse response
h = np.zeros(101)
for i in range(20):
  h[i+40] = random.randint(0,5)

# signal
x = np.zeros(101)

for i in range(20):
  x[i+45] = random.randint(0,5)

t = np.arange(-50, 51, 1)
```

### Impulse Response

```python
plt.figure(figsize=(25,5))
plt.title('Impulse Response')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, h)
```

### Signal

```python
plt.figure(figsize=(25,5))
plt.title('Input Signal')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, x)
```

## Convolution

### Implementation

```python
# convolution

y = []

h_inv = h[::-1]

for i in t:
  sum = 0
  for k in t:
    sum += x[k]*h_inv[k-i]
  y.append(sum)

plt.figure(figsize=(25,5))
plt.title('Convolution')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, y)
```

### NumPy

```python
np_y = np.convolve(x, h, 'same')
print(len(np_y))
plt.figure(figsize=(25,5))
plt.title('Convolution with Numpy')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, np_y)
```
