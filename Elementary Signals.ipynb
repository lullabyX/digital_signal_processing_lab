# 01 Generate and plot the elementary signals (unit impulse, unit step, ramp and exponential).

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Necessary functions
'''
UL - Upper Limit
LL - Lower Limit
t  - time
'''

UL = 81
LL = -80

t = np.arange(LL, UL , 1)

total = UL + abs(LL)
mid = math.ceil(total/2);

lim = 1.0

# Delay function
def Delay(x_n, k):
  
  ret = np.zeros(abs(k))
  if(k>=0):
    ret = np.concatenate((ret, x_n[0:len(x_n)-k]))
  else:
    ret = np.concatenate((x_n[abs(k):len(x_n)], ret))

  return ret
```

## Unit Impulse Signal

```python
# delay
k = -4

# Unit Impulse
unit_impulse = np.zeros(total)
unit_impulse[mid] = 1
unit_impulse = Delay(unit_impulse, k)

plt.figure(figsize=(25,5))
plt.title('Unit Impulse Signal')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, unit_impulse)
```

## Unit Step

```python
# delay
k = 3

unit_step = np.zeros(total)
unit_step[mid:] = 1
unit_step = Delay(unit_step, k)

plt.figure(figsize=(25,5))
plt.title('Unit Step Signal')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, unit_step)
```

## Unit Ramp

```python
# delay
k = 0

unit_ramp = np.zeros(total)
for x in range(mid, total):
  unit_ramp[x] = x-mid+1
unit_ramp = Delay(unit_ramp, k)

plt.figure(figsize=(25,5))
plt.title('Unit Ramp Signal')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, unit_ramp)
```

## Exponential Signal

```python
# delay
k = 0

t = np.arange(-1, 1, 0.01)
expo =[]
for x in t:
  expo.append(np.exp(x*2))

expo = Delay(expo, k)

plt.figure(figsize=(25,5))
plt.title('Exponential Signal')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.stem(t, expo)
```
