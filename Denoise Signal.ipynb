# 04 Compute the inverse DFT of the signal obtained in (3).

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

## Generating Previous DFT

```python
t = np.arange(0, 2, 0.001)

# signal 1
x1 = 2 * np.sin(2*np.pi*3*t)

# signal 2

x2 = 3 * np.sin(2 * np.pi * 5 * t)

# signal 3

x3 = 4 * np.sin(2 * np.pi * 7 * t)

# Combined Signal

x = x1 + x2 + x3

plt.figure(figsize=(25,3))
plt.title('Combined Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.plot(t, x, color='purple')
```

```python
def DFT(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    
    X = np.dot(e, x)
    
    return X
```

```python
dft = DFT(x)

plt.figure(figsize = (25,5))
plt.title('Previous DFT(in complex domain)')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.plot(t, dft)
```

## Inverse DFT

```python
def inverse_dft(x): 
  N = len(x)
  n = np.arange(0, N)
  k = n.reshape((N, 1))
  e = np.exp(2j * np.pi * k * n / N)

  X = np.dot(e, x)/N
    
  return X
```

```python
X = inverse_dft(dft)

plt.figure(figsize = (25,5))
plt.xlabel('Amplitude')
plt.ylabel('Time')
plt.axvline(linewidth = '1', color = 'r')
plt.axhline(linewidth = '1', color = 'r')
plt.plot(t,X, color='green')
```
